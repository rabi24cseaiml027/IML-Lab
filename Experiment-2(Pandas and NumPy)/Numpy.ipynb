```python
import numpy as np

a1 = np.array([[1, 2, 3],[4, 5, 6]])

a2 = np.array([[1, 2],[3, 4],[5, 6]])

a3 = np.zeros((2, 2))

for i in range(2):
    for j in range(2):
        for k in range(3):
            a3[i][j] += a1[i][k] * a2[k][j]

print(a3)
```

```python
a1 = [1,2,3,4,5,6]
print('Maximum is', np.max(a1))
print("Mininmum is", np.min(a1))
print("Average maximum is", np.argmax(a1))
print("Average minimum is", np.argmin(a1))
```

```python
a1 = np.array([1, 2, 3])
a2 = np.array([4, 5, 6])
print(np.concatenate((a1, a2)))
print(np.stack((a1, a2)))
print(np.vstack((a1, a2)))
```

```python
a1 = np.array([1, 2, 3])
a2 = np.array([4, 5, 6])
print("Element wise equal", a1 == a2)
print("Element wise greater", a1 > a2)
print("Element wise less", a1 < a2)
```
